```python
# compile the model 
model.compile(
    optimizer = 'adam',loss = 'mean_squared_error'
)

# train the model 

model.fit(X_train,y_train,epochs=50,batch_size = 32,verbose=1)

# evaluate the model 
loss = model.evaluate(X_test,y_test)

print('Test Loss',loss)
```

```python
# Step 1 import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from keras import layers

# import from scikit-learn library

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler,LabelEncoder

# import datasets 
df = sns.load_dataset('tips')

# preprocess the data
le = LabelEncoder()
df['sex'] = le.fit_transform(df['sex'])
df['smoker'] = le.fit_transform(df['smoker'])
df['day'] = le.fit_transform(df['day'])
df['time'] = le.fit_transform(df['time'])

# convert tip into binnary variable 
df['tip'] = (df['tip']> 2).astype(int)


X = df.drop('tip',axis=1)
y = df['tip']

# Scaled the data for better result
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Train Test split
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=42)

# Create the model 
model = tf.keras.Sequential([
    tf.keras.layers.Dense(32, activation='relu',input_shape=[X_train.shape[1]]), # input layer
    tf.keras.layers.Dense(16, activation='relu'), # hidden layer
    tf.keras.layers.Dense(1,activation='sigmoid')
])

# compile the model 
model.compile(
    optimizer = 'adam',loss = 'binary_crossentropy',metrics=['accuracy']
)

# train the model 

model.fit(X_train,y_train,epochs=50,batch_size = 32,verbose=1)

# evaluate the model 
loss ,accuracy= model.evaluate(X_test,y_test)

print('Test Loss',loss)
print('Test Accuracy',accuracy)
```
